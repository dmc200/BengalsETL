```python
bengals_df = pd.read_csv('s3://mindex-data-analytics-code-challenge/bengals.csv')
boyd_df = pd.read_csv('s3://mindex-data-analytics-code-challenge/boyd_receiving.csv')
chase_df = pd.read_csv('s3://mindex-data-analytics-code-challenge/chase_receiving.csv')
higgins_df = pd.read_csv('s3://mindex-data-analytics-code-challenge/higgins_receiving.csv')
```

```python

bengals_df['Result'] = bengals_df['Result'].apply(lambda x: 'Win' if x == 1.0 else 'Loss')
bengals_df
```

```python
chase_df= chase_df.rename(columns={'Yards': 'Chase Yards', 'TD': 'Chase TD'})
boyd_df= boyd_df.rename(columns={'Yards': 'Boyd Yards', 'TD': 'Boyd TD'})
higgins_df= higgins_df.rename(columns={'Yards': 'Higgins Yards', 'TD': 'Higgins TD'})
```

```python
merged_df = pd.merge(bengals_df, chase_df, on='Week', how='left') 
m2 = pd.merge(merged_df,boyd_df, on='Week', how='left') 
m3 = pd.merge(m2, higgins_df, on='Week', how='left') 

m3
```
